```python
#Importing important libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatter plot.
#Creating data
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,100,86,103,87,94,78,77,85,86]
#Setting figure size in Inches
plt.figure(figsize=(5,3))
#Plotting the scatter plot
plt.scatter(x,y,color='red')
#Set the x and y labels
plt.xlabel ('X Values')
plt.ylabel ('Y Values')
#Set the title
plt.title ('Scatter Plot')
#Show the plot
plt.show()
#
```

```python
#Line plot
#Creating the data
year=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
unemployment_rate=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3,9.9]
#Plot
plt.figure(figsize=(5,3))
plt.plot(year,unemployment_rate, color='green',marker='o')
plt.title('Unemployment Rate across Years')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
#Bar Chart
#Create the data
data={
    'C':20,
    'C++':15,
    'Java':30,
    'Python':35
}

#Setting the figure size
plt.figure(figsize=(5,3))
#Plotting the bar chart
plt.bar(data.keys(),data.values(),color='black',width= 0.6 )
#Set the labels
plt.xlabel('Programming Languages')
plt.ylabel('Number of students')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
#Loading inbuilt dataset from seaboarn
df = sns.load_dataset('tips')
df.head()
```

```python
#Doubt
print(df.info())
print(df.shape)
```

```python
#countplot
plt.figure(figsize=(5,3))
sns.countplot(data=df,x='day', color='green', width=0.6)
plt.show()

```

```python
#countplot
plt.figure(figsize=(5,3))
sns.countplot(data=df,x='smoker', color='red')
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(5,3))
plt.boxplot(df.total_bill)
plt.title('Boxplot of Total Bill')
plt.ylabel('Total Bill')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(5,3))
plt.hist(df.total_bill,bins=30,color='green')
plt.title('Histogram of Total Bill')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
#
```

```python
#Stacked Bar Chart
#Creating Data
import numpy as np
x=['A','B','C','D']
y1=np.array([10,20,10,30])
y2=np.array([20,2,15,25])
y3=np.array([12,15,19,6])
y4=np.array([10,25,15,20])

#Bar chart in stack manner
plt.bar(x,y1, color='green')
plt.bar(x,y2, bottom= y1, color='red')
plt.bar(x,y3, bottom=y1+y2, color='yellow')
plt.bar(x,y4, bottom=y1+y2+y3 , color='blue')

#set X and Y labels
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Stack Bar chart')
plt.legend(['Round 1','Round 2','Round 3','Round 4'])
plt.show()




```
